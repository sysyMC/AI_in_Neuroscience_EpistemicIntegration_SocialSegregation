# Temporal intrinsic AI activity of the journals (computed with all their respective publications and not only the most impactful ones)

```python
# equivalent to level 0 here
ids_neuro = pd.read_csv(r"/home/sylvain/DOCTORAT/DATA/microsoft_academic/corpus/neuroscience/database/updated/QL0/idsNeuroQl0.csv",
                        sep = ';')
```

```python
allneuropapers = pd.read_csv(allneuropapers_file, sep = ';')
allneuropapers = allneuropapers[(allneuropapers['publicationYear']>=1970) 
                                & (allneuropapers['publicationYear']<=2019)
                                & (allneuropapers['id'].isin(ids_neuro['id']))].explode('journal_title')
allneuropapers['journal_title'] = [list(set(eval(i)))[0] if i!='-1' else -1 for i in allneuropapers['journal_title']]
allneuropapers = allneuropapers[allneuropapers['journal_title']!=-1]
allneuropapers
```

```python
# the activities are defined until 2019
def activity_journal(journal,df):
    activity = []
    for y in range(1970,2021):
        AIscore_publications = df[(df['journal_title']==journal) & (df['publicationYear']==y)]['isIA']
        if len(AIscore_publications) == 0:
            activity.append(-1)
        else:
            activity.append(AIscore_publications.mean())
    return activity

def all_activities_journals(df):
    total_activity_journal = {}
    temporal_activity_journal = {}
    for journal in tqdm(set(df['journal_title'])-{-1}):
        total_activity_journal[journal] = df[df['journal_title']==journal]['isIA'].mean()
        temporal_activity_journal[journal] = activity_journal(journal,df)
    return total_activity_journal, temporal_activity_journal

def total_activities_journals(df):
    total_activity_journal = {journal: df[df['journal_title']==journal]['isIA'].mean() 
                              for journal in tqdm(set(df['journal_title'])-{-1})}
    return total_activity_journal
```

```python
total_activity_journal,temporal_activity_journal = all_activities_journals(allneuropapers)
```

### Clean temporal activities of journals with too much nan values after their first issue
Two main cases encountered in our database:
- Stopped activity during a long time period. For instance, according to Scopus, the journal 'PROGRESS IN NEUROLOGY AND PSYCHIATRY' has an activity since 1948 until 1972, then stopped until 2007. Unless it is related to two different journals but with same name, the ISSN seems unique in OpenAlex and Scopus. It could be also the case of the same journal with same ISSN over time but with different names according to its history (changing editorial board, scope, etc.). In all these cases, let them with their own nan values because it reflects the history of the journal. If the activity is not significant over time, we delete it.
- For some other journals with an activity in 1970 and nothing during more than 15 or 20 years, especially if they have an important AI activity, we correct them manually according to their history grabbed on Wikipedia or on major publishers' websites (Springer, Nature, SAGE, etc.). Indeed, the papers apparently published in 1970 in such journals are often effectively published later after some requests on APIs of scientometric databases (Web of Science, Scopus or OpenAlex). This could be the result of some mistakes during the grabbing done by MAG. This is the case of 'FRONTIERS IN NEUROINFORMATICS' below.

```python
begin_year = [np.where(np.array(temporal_activity_journal[j])!=-1)[0][0] for j in temporal_activity_journal.keys()]
list_journals = [list(temporal_activity_journal.keys())[k] for k in np.argsort(begin_year)]
temporal_activity_journal_ranked = {k:temporal_activity_journal[k] for k in list_journals}
total_activity_journal_ranked = {k:total_activity_journal[k] for k in list_journals}
matrix = pd.DataFrame(temporal_activity_journal_ranked).replace(-1,np.nan).values[:-1]
```

```python
messyJournals = []
j = 0
for aj in matrix.transpose():
    by = np.where(~np.isnan(aj))[0][0]
    if len(np.where(np.isnan(aj[by:]))[0])>(50-by)/4:
        messyJournals.append((j,list_journals[j]))
    j+=1
```

## Relation between AI score of the journal and of the authors publishing in it (1970-2019)

```python
AI_score_journals = allneuropapers[['journal_title','isIA']].groupby('journal_title').mean('isIA').reset_index()
AI_score_journals = AI_score_journals[AI_score_journals['journal_title'].isin(list_journals)]
AI_score_journals
```

```python
authors_journals = allpapers[['journal_title','authors']].reset_index(drop=True)
authors_journals['authors'] = [list(set(i)) if i!=-1 else -1 for i in authors_journals['authors']] # to avoid the double apparitions
authors_journals = authors_journals.explode('authors')
authors_journals['counter'] = 1
authors_journals = authors_journals.groupby(['authors','journal_title']).sum().reset_index() #reset_index transform all index as columns by default
authors_journals = authors_journals[(authors_journals['authors']!=-1)
                                    & (authors_journals['journal_title'].isin(list_journals))]
authors_journals.rename(columns={'authors':'authorId'}, inplace=True)
authors_journals
```

```python
authors_journals = pd.merge(authors_journals, authorsInPapers[['authorId','AIscore_all']], on='authorId', how='inner')
authors_journals = pd.merge(authors_journals, AI_score_journals, on='journal_title', how='left')
authors_journals
```

```python
# compute the mean AI score of the authors per journal
couples_mean = []
std = []
sem = []
for journal in tqdm(list_journals):
    scores_authors = authors_journals[authors_journals['journal_title']==journal]['AIscore_all']
    couples_mean.append((AI_score_journals[AI_score_journals['journal_title']==journal]['isIA'].values[0],
                         scores_authors.mean()))
    std.append(scores_authors.std())
    sem.append(scores_authors.sem())
```

```python
colormap = plt.cm.copper
colors = [colormap(i) for i in np.linspace(min(std),1,len(std))]
```

```python
fig = plt.figure(figsize=(7, 3))
ax = plt.axes()
'''
ax.errorbar([x[0] for x in couples_mean],
        [x[1] for x in couples_mean], 
        yerr = sem,
        fmt='o', 
        mfc='none',
        elinewidth=0.5,
        capsize=3,
        markersize=7, 
        color=plt.cm.tab20.colors[0],
        alpha=0.7)    
        #'o', mfc='none', markersize=7, color=plt.cm.tab20.colors[0], alpha=0.7)
'''
g = ax.scatter([x[0] for x in couples_mean],
        [x[1] for x in couples_mean], 
        marker='o', 
        facecolor='none',
        s=50, 
        edgecolors=colors,
        alpha=0.5)   
ax.set_ylabel(r'$\overline{f_{AI}}$', fontsize=axisLabelSize)
ax.set_xlabel(r'$a_{AI}$', fontsize=axisLabelSize)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim(0,1)
ax.set_xlim(0,1)
cbar = ax.figure.colorbar(mappable=plt.cm.ScalarMappable(cmap='copper'),ax=ax)
cbar.ax.set_ylabel(r'$\sigma_{f_{AI}}$', rotation=-90, va='bottom', fontsize=20)
cbar.ax.set_ylim(0,1)
cbar.ax.set_yticks(np.linspace(0,1,3),np.linspace(0,0.25,3))
cbar.ax.tick_params(which='major', labelsize=ticksLabelSize, length=ticksSizeMajor-3, width=ticksWidth)
ax.tick_params(which='major', labelsize=ticksLabelSize, length=ticksSizeMajor, width=ticksWidth)
ax.tick_params(which='minor', labelsize=ticksLabelSize, length=ticksSizeMinor, width=ticksWidth)
plt.savefig(f'{pics_folder}correlation_AIscore_activity_log.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
